```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'D:\ShareData\PythonFinance\walmart_stock.csv',index_col='Date',parse_dates=True).sort_values(by='Date', ascending=True)
```

```python
df['Open'].plot(figsize=(16,6))
```

```python
# 7 day rolling mean
df.rolling(7).mean().head(20) #Moving avg of 7 days.
```

```python
df['Close'].plot()
df.rolling(window=30).mean()['Close'].plot()
```

```python
#Adding Legend
df['Close: 30 Day Mean'] = df['Close'].rolling(window=30).mean()
df[['Close','Close: 30 Day Mean']].plot(figsize=(16,6))
```
